```python
import requests
from bs4 import BeautifulSoup
# Base URL for Jumia search results
base_url = "https://www.jumia.co.ke/catalog/?q=apple+smartphone&page="

# Loop through pages 1 to 8
for i in range(1, 9):
    url = base_url + str(i)
    response = requests.get(url)
    
    if response.status_code == 200:
        print(f"Scraping: {url}")  # Debugging output to confirm pagination
        soup = BeautifulSoup(response.text, "lxml")

        # Extract product names and prices
        products = soup.find_all("article", class_="prd")

        for product in products:
            title = product.find("h3", class_="name")
            price = product.find("div", class_="prc")

            if title and price:
                print(f"Product: {title.text.strip()} - Price: {price.text.strip()}")
    
    else:
        print(f"Failed to fetch page {i}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

# Output CSV file
filename = "jumia_apple_products.csv"

# Base URL for Jumia search results
base_url = "https://www.jumia.co.ke/catalog/?q=apple+smartphone&page="

# Open CSV file for writing
with open(filename, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Product Name", "Price"])  # CSV Header

    # Loop through pages 1 to 8
    for i in range(1, 9):
        url = base_url + str(i)
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})  # Avoid bot detection

        if response.status_code == 200:
            print(f"Scraping: {url}")  # Debugging output
            soup = BeautifulSoup(response.text, "lxml")

            # Extract product names and prices
            products = soup.find_all("article", class_="prd")

            for product in products:
                title = product.find("h3", class_="name")
                price = product.find("div", class_="prc")

                if title and price:
                    writer.writerow([title.text.strip(), price.text.strip()])  # Save to CSV
                    print(f"Product: {title.text.strip()} - Price: {price.text.strip()}")  # Display output

        else:
            print(f"Failed to fetch page {i}")

        time.sleep(2)  # Delay to prevent getting blocked

print(f"\nScraped data saved to {filename}")
```
